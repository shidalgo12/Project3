```python
import sklearn

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
# File to Load
file_to_load = "NBA_Averages_Per_Season.csv"

# Read .csv File and store into Pandas data frame
averages_df = pd.read_csv(file_to_load)

averages_df.head()
```

```python
seasons_df = averages_df[['Season', 'Age','PTS','3P%','FT/FGA', 'ORtg']]
seasons_df.head()
```

```python
seasons_df.boxplot(column=['Age', 'PTS', '3P%', 'FT/FGA'], grid=False)
```

```python
seasons_df_t = seasons_df.T

seasons_df_t
```

```python
header_row = 0

seasons_df_t.columns = seasons_df_t.iloc[header_row]
seasons_df_t.head()
```

```python
seasons_df_t = seasons_df_t.iloc[1:]
seasons_df_t
```

```python
era1_df = seasons_df_t[['1984-85', '1985-86','1986-87','1987-88','1988-89']]
era1_df.head()
```

```python
era1_df_t = era1_df.T
era1_df_t
```

```python
era1_df_t.mean(axis=0)
era1_df_t
```
